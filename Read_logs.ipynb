```python
# This part of the code reads the log files (containing maven logs) of Log4J project
# And converts the logs to CSV format (Test_case, Result)


import pandas as pd
import os


for_fold = 'fold10'
fold = 'logData/'+for_fold+'/output_Execution'
all_files_in_fold = os.listdir(fold+'/')

##
my_set = {}

# combined-result data-frame

new_df = pd.DataFrame( columns=['file_name', 'Result'])

# previous key
prev_key =""

#PAth to CSV Files
path = 'logData/apfdAll/'+for_fold
os.makedirs(path)
for file_name in all_files_in_fold:
    reading_file = open(fold+'/'+file_name, "r")
    for line in reading_file:
            if "Test" in line or "Errors:" in line:
                if "at " not in line and "log4j:" not in line:
                    #print(line)
                    if "Running" in line: # Name of Test Case (Key for Dict)
                        prev_key = line.split()[1]
                    if "Tests run:" in line: # Test Case Result (Value for Dict)
                        if "FAILURE!" in line:
                            my_set[prev_key] = "failed" # Test Case Failed
                        else:
                            my_set[prev_key] = "passed" #Test Case Passed
    tmp_df = pd.DataFrame(my_set.items(), columns=["file_name", 'Result'])
    tmp_df.to_csv(path+'/'+file_name.split('.')[0])
    
```
